# K-means

````python
from sklearn.cluster import KMeans

# k-means
x = []   # k 가 몇 개인지 
y = []   # 응집도가 몇인지 

for k in range(1, 11):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(c)
    
    x.append(k)
    y.append(kmeans.inertia_)
````

Inertia value는 군집화가 된 후에, 각 중심점에서 군집의 데이터 간의 거리를 합산한 것이으로 군집의 응집도를 나타내는 값이다, 이 값이 작을수록 응집도가 높게 군집화가 잘되었다고 평가할 수 있다.

````python
plt.plot(x, y)
````

군집수가 많아질수록 Inertia value가 낮아지는데, 큰 폭으로 감소하는 2-4가 적당해보임

````python
from sklearn.metrics import silhouette_samples, silhouette_score
````

````python
params = {'n_clusters':[2, 3, 4, 5, 6, 7, 8, 9, 10]}

scores_kmeans = {}
for n_cluster in params['n_clusters']:
    model = KMeans(n_clusters=n_cluster)
    model.fit(c.iloc[:,:3])
    score = silhouette_score(c.iloc[:,:3], model.labels_)
    scores_kmeans['='.join(['n', str(n_cluster)])] = score
    
pd.Series(scores_kmeans).sort_values(ascending=False)[:10]
````

*inertia는 4가 더 낮은데 실루엣은 2일 때 더 높게 나옴 뭐지*

데이터 분포를 보고 n=2 또는 n=4로 정하면 될 것 같다.

````python
from sklearn.cluster import AgglomerativeClustering
### 여러개의 클러스터링 갯수를 List로 입력 받아 각각의 실루엣 계수를 면적으로 시각화한 함수 작성
def visualize_silhouette(clustering, X_features, cluster_lists=[1]): 
    
    # 입력값으로 클러스터링 갯수들을 리스트로 받아서, 각 갯수별로 클러스터링을 적용하고 실루엣 개수를 구함
    n_cols = len(cluster_lists)
    
    # plt.subplots()으로 리스트에 기재된 클러스터링 수만큼의 sub figures를 가지는 axs 생성 
    fig, axs = plt.subplots(figsize=(4*n_cols, 4), nrows=1, ncols=n_cols)
    
    # 리스트에 기재된 클러스터링 갯수들을 차례로 iteration 수행하면서 실루엣 개수 시각화
    for ind, n_cluster in enumerate(cluster_lists):
        
        if clustering[0] == 'hierarchical':
            model = AgglomerativeClustering(n_clusters=n_cluster, linkage=clustering[1])
        elif clustering[0] == 'kmeans':
            model = KMeans(n_clusters=n_cluster)
        # KMeans 클러스터링 수행하고, 실루엣 스코어와 개별 데이터의 실루엣 값 계산. 
        cluster_labels = model.fit_predict(X_features)
        
        #if clustering[0] == 'dbscan':
        #    n_cluster = len(set(cluster_labels))-1
        
        sil_avg = silhouette_score(X_features, cluster_labels)
        sil_values = silhouette_samples(X_features, cluster_labels)
        
        y_lower = 10
        axs[ind].set_title('Number of Cluster : '+ str(n_cluster)+'\n' \
                          'Silhouette Score :' + str(round(sil_avg,3)) )
        axs[ind].set_xlabel("The silhouette coefficient values")
        axs[ind].set_ylabel("Cluster label")
        axs[ind].set_xlim([-0.1, 1])
        axs[ind].set_ylim([-1, len(X_features) + (n_cluster + 1) * 10])
        axs[ind].set_yticks([])  # Clear the yaxis labels / ticks
        axs[ind].set_xticks([0, 0.2, 0.4, 0.6, 0.8, 1])
        
        # 클러스터링 갯수별로 fill_betweenx( )형태의 막대 그래프 표현. 
        for i in range(n_cluster):
            ith_cluster_sil_values = sil_values[cluster_labels==i]
            ith_cluster_sil_values.sort()
            
            size_cluster_i = ith_cluster_sil_values.shape[0]
            y_upper = y_lower + size_cluster_i
            
            color = cm.nipy_spectral(float(i) / n_cluster)
            axs[ind].fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_sil_values, \
                                facecolor=color, edgecolor=color, alpha=0.7)
            axs[ind].text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
            y_lower = y_upper + 10
            
        axs[ind].axvline(x=sil_avg, color="red", linestyle="--")
````

# GMM

````python
from sklearn.mixture import GaussianMixture

n_components = np.arange(1, 11)
models = [GaussianMixture(n, covariance_type='full', random_state=42).fit(c.iloc[:,:3]) for n in n_components]
plt.plot(n_components, [m.bic(c.iloc[:,:3]) for m in models], label='BIC')
plt.plot(n_components, [m.aic(c.iloc[:,:3]) for m in models], label='AIC')
plt.legend(loc='best')
plt.xlabel('n_components');
````

n=2, n=4에서 낮은 AIC, BIC 값을 가짐

````python
gmm = GaussianMixture(n_components=4, random_state=1107)
gmm_labels = gmm.fit_predict(c.iloc[:,:3])

c['gmm_cluster'] = gmm_labels

# print(data_pca.groupby('키즈카페수')['gmm_cluster'].value_counts())
````

# AffinityPropagation

```
DBSCAN, Spectral Clustering도 해봤는데.. 
밀도 기반 군집화 방법이라 우리 데이터는 밀도가 낮아서 전체 데이터가 모두 같은 그룹으로 형성됨
-> ep 값을 아주 크게 하면 군집이 형성되긴 한다고 한다
```

**Affinity Propagation은 데이터 쌍 간의 유사도를 이용하여 데이터를 군집화**

- 클러스터의 수를 사전에 지정할 필요가 없으며, 클러스터의 중심점을 자동으로 정하므로 군집 개수를 추정하는데 유용

- 대부분의 군집화 알고리즘과는 달리 군집의 형태가 복잡하거나, 군집 내 데이터의 개수가 다른 경우에도 잘 작동하는 특징

- 중소 규모의 데이터셋에 적합한 군집화 알고리즘

````python
from sklearn.cluster import AffinityPropagation

clustering = AffinityPropagation(random_state=1107).fit(c.iloc[:,:3])
c['aff_cluster'] = clustering.labels_
AffinityPropagation(random_state=1107)
````

````python
plt.figure(figsize=(10, 7))
plt.scatter(c.iloc[:,1], c.iloc[:,2], c=c['aff_cluster'], cmap='rainbow')
````
